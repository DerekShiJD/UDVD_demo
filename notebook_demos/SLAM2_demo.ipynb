```python
# denoise entire video

model.eval()

frames = []
os.makedirs(os.path.join("../examples", video), exist_ok=True)
save_dir = os.path.join("../examples", video)
C, H, W = test_loader.dataset[0].shape

# static
# x = 0; y = 0; w = 1024; h = 1024

# output frame list
noisy_list = []
clean_list = []

with torch.no_grad():
    for i, noisy_inputs in enumerate(test_loader):

        noisy_inputs = noisy_inputs[:,:,:,:].to(device)
        
        N, C, H, W = noisy_inputs.shape
        # print(N,C,H,W)   # 1 5 1024 1024
        noisy_frame = noisy_inputs[:, 2:3, :, :]
        
        output, _ = model(noisy_inputs)
        
        #img = np.array((output[0]/max(1,output[0].max())).cpu().detach().numpy().reshape(w,h)*255, dtype=np.uint8)
        #noisy_img = np.array((noisy_frame[0]/max(1,noisy_frame[0].max())).cpu().detach().numpy().reshape(w,h)*255, dtype=np.uint8)
        
        img = output[0].cpu().detach().numpy().reshape(H,W)
        noisy_img = noisy_frame[0].cpu().detach().numpy().reshape(H,W)
        
        
        # save to tiff
        temp_noisy = noisy_img.reshape((1, H, W))
        temp_clean = img.reshape((1, H, W))
        
        temp_clean[temp_clean<0] = 0
        
        noisy_list.append(temp_noisy)
        clean_list.append(temp_clean)
        
        # save to jpg (<=255)
        
        img = img / 0.1
        img[img>1] = 1
        img[img<0] = 0
        img = np.array(img * 255, dtype=np.uint8)
        
        noisy_img = noisy_img / 0.1
        noisy_img[noisy_img>1] = 1
        noisy_img[noisy_img<0] = 0
        noisy_img = np.array(noisy_img * 255, dtype=np.uint8)
        
        frame = cv2.cvtColor(np.append(noisy_img, img, axis=1), cv2.COLOR_GRAY2BGR)
        cv2.line(frame, (W,0), (W,H), (0,0,0), 1)
        cv2.imwrite(os.path.join(save_dir, "%05d.jpg" % (i)), frame)
        frames.append(frame)
        # break
    
    
# save to tiff
img_clean = np.concatenate(clean_list, axis=0)
img_noisy = np.concatenate(noisy_list, axis=0)
tifffile.imwrite(os.path.join(save_dir, 'Image_output.tif'), img_clean)
tifffile.imwrite(os.path.join(save_dir, 'Image_input.tif'), img_noisy)


```
